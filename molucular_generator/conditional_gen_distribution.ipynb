```python
given_mw = [250, 300, 350, 400, 450]

generated_mw_values = {
    250: mw_1['MW'],
    300: mw_2['MW'],
    350: mw_3['MW'],
    400: mw_4['MW'],
    450: mw_5['MW'],
}

fig, ax = plt.subplots(figsize=(8, 6))

means = [np.mean(generated_mw_values[mw]) for mw in given_mw]
stds = [np.std(generated_mw_values[mw]) for mw in given_mw]
ax.errorbar(given_mw, means, yerr=stds, fmt='o', capsize=5, label='Mean and Standard Deviation')

parts = ax.violinplot(list(generated_mw_values.values()), positions=given_mw, widths = 25, showmeans=False, showextrema=False, showmedians=True)
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_alpha(0.6)

ax.plot([min(given_mw), max(given_mw)], [min(given_mw), max(given_mw)], color='dimgray', linestyle='--')

plt.xticks(given_mw, fontsize=20)
plt.yticks(fontsize=20)

matplotlib.rcParams.update({'font.size': 20})
plt.title('Distrubution - MW', fontsize=26)
plt.xlabel('Given MW', fontsize=20)
plt.ylabel('Generated MW', fontsize=20)

legend_handles, legend_labels = ax.get_legend_handles_labels()
legend_handles.append(plt.Line2D([0], [0], color='orange', lw=2, label='Median'))
ax.legend(handles=legend_handles, loc='lower right', fontsize=20)

plt.grid(True)
plt.show()
```
